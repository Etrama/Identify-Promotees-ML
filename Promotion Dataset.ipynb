# Splitting Data into train and test set

```python
Y = target
X = new_df7.drop(columns='employee_id')
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 1)
```

# Extra Trees CLassifier

```python
from sklearn.ensemble import ExtraTreesClassifier
#NUmber of cores used to perform parallel fitting of the foret model
n_jobs = 4
print("Fitting ExtraTreesCLassifier on data with %d cores..." % n_jobs)
t0 = time()
ETC = ExtraTreesClassifier(n_estimators = 100,
                          n_jobs = n_jobs,
                          random_state=0)
ETC.fit(X_train, Y_train.values.ravel())
print("done in %0.3fs" % (time() - t0))
importances = ETC.feature_importances_
```

```python
print("Accuracy Score is: {}".format(ETC.score(X_test, Y_test)))
print(importances)
```
